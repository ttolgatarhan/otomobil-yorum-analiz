```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import csv
from zemberek import TurkishMorphology
from zemberek import TurkishSentenceNormalizer

morphology = TurkishMorphology.create_with_defaults()
normalizer = TurkishSentenceNormalizer(morphology)

options = Options()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(options=options)

driver.get("https://eksisozluk.com/basliklar--1208965?nr=true&rf=basliklar")
time.sleep(5)

with open("arac.txt", "r", encoding="utf-8") as file:
    arac_listesi = [line.strip() for line in file]

with open('entries.csv', mode='w', newline='', encoding='utf-8') as csv_file:
    fieldnames = ['Araç', 'Sayfa', 'Başlık', 'Entry No', 'Entry Metni', 'Lemmatize Entry Metni']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()

    for arac in arac_listesi:
        print(f"\n🔍 Araç: {arac}")

        try:
            print("⏳ Arama kutusuna araç ismi yazılıyor...")
            input_btn = driver.find_element(By.XPATH, '/html/body/header/div/form[1]/input')
            input_btn.clear()
            input_btn.send_keys(arac)

            search_btn = driver.find_element(By.XPATH, '/html/body/header/div/form[1]/button')
            search_btn.click()
            time.sleep(5)

            current_url = driver.current_url
            print("🌐 URL:", current_url)

            for page_number in range(2, 100):
                print(f"📄 Sayfa {page_number} yükleniyor...")
                driver.get(f"{current_url}?p={page_number}")
                page_source = driver.page_source
                soup = BeautifulSoup(page_source, "html.parser")

                try:
                    baslik = soup.find("h1").text.strip()
                    print(f"\n📌 Başlık: {baslik}")
                except Exception as e:
                    print(f"Başlık bulunamadı. Hata: {e}")

                entries = soup.select("ul li div.content")
                if not entries:
                    print("⚠️ Entry bulunamadı.")
                    continue

                for idx, entry in enumerate(entries, 1):
                    entry_metni = entry.text.strip()
                    try:
                        lemmatized_entry = normalizer.normalize(entry_metni)
                    except Exception as e:
                        lemmatized_entry = entry_metni
                        print(f"⚠️ Lemmatize hatası: {e}")

                    writer.writerow({
                        'Araç': arac,
                        'Sayfa': page_number,
                        'Başlık': baslik,
                        'Entry No': idx,
                        'Entry Metni': entry_metni,
                        'Lemmatize Entry Metni': lemmatized_entry
                    })
                    print(f"\n💬 {idx}. Entry:\n{entry_metni}\nLemmatize Edilmiş:\n{lemmatized_entry}\n{'-' * 40}")

        except Exception as e:
            print("🚫 Hata:", e)

        time.sleep(2)

driver.quit()
```

```python
import pandas as pd
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')

def tokenize_dataset(path):
    df = pd.read_csv(path).dropna()
    df.columns = ['metin']
    return df['metin'].apply(lambda x: word_tokenize(str(x).lower(), language='turkish')).tolist()

def train_and_save(sentences, model_type, window, vector_size, label):
    sg = 1 if model_type == 'skipgram' else 0
    model = Word2Vec(sentences=sentences, vector_size=vector_size, window=window, sg=sg, min_count=1)
    filename = f"word2vec_{label}_sg{sg}_w{window}_dim{vector_size}.model"
    model.save(filename)
    print(f"✅ Kaydedildi: {filename}")

sentences_lem = tokenize_dataset("lemmatized.csv")
sentences_stem = tokenize_dataset("stemmed.csv")

parameters = [
    {'model_type': 'cbow', 'window': 2, 'vector_size': 100},
    {'model_type': 'skipgram', 'window': 2, 'vector_size': 100},
    {'model_type': 'cbow', 'window': 4, 'vector_size': 100},
    {'model_type': 'skipgram', 'window': 4, 'vector_size': 100},
    {'model_type': 'cbow', 'window': 2, 'vector_size': 300},
    {'model_type': 'skipgram', 'window': 2, 'vector_size': 300},
    {'model_type': 'cbow', 'window': 4, 'vector_size': 300},
    {'model_type': 'skipgram', 'window': 4, 'vector_size': 300}
]

for params in parameters:
    train_and_save(
        sentences_lem,
        model_type=params['model_type'],
        window=params['window'],
        vector_size=params['vector_size'],
        label=f"lemmatized"
    )
    train_and_save(
        sentences_stem,
        model_type=params['model_type'],
        window=params['window'],
        vector_size=params['vector_size'],
        label=f"stemmed"
    )
```

```python
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
import os

def average_vector(tokens, model):
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(model.vector_size)

giris_cumlesi = "enzer motor donanımdaki linealardan tl pahalı olacağı söyleniyor kadar doğrudur bilemem"
giris_tokens = giris_cumlesi.lower().split()

model_files = [f for f in os.listdir() if f.endswith(".model")]

result_list = []

for model_file in sorted(model_files):
    try:
        model = Word2Vec.load(model_file)

        if "lemmatized" in model_file:
            df = pd.read_csv("lemmatized.csv")
        elif "stemmed" in model_file:
            df = pd.read_csv("stemmed.csv")
        else:
            continue  

        df.columns = ['metin']
        sentences = df['metin'].astype(str).apply(lambda x: x.lower().split()).tolist()

        giris_vector = average_vector(giris_tokens, model)
        all_vectors = [average_vector(s, model) for s in sentences]
        sim_scores = cosine_similarity([giris_vector], all_vectors)[0]
        top5 = sim_scores.argsort()[::-1][0:5]  

        for i in top5:
            result_list.append({
                "model": model_file,
                "benzerlik_skoru": sim_scores[i],
                "benzer_cumle": " ".join(sentences[i])
            })

    except Exception as e:
        print(f"❌ HATA {model_file}: {str(e)}")

pd.DataFrame(result_list).to_csv("word2vec_benzerlik_sonuclari.csv", index=False)
print("✅ Sonuçlar kaydedildi: word2vec_benzerlik_sonuclari.csv")
```

```python
import pandas as pd

giris_cumlesi = "benzer motor donanımdaki linealardan tl pahalı olacağı söyleniyor kadar doğrudur bilemem."

def tokenize(text):
    return set(str(text).lower().split())

def jaccard_similarity(a, b):
    a_set = tokenize(a)
    b_set = tokenize(b)
    if not a_set or not b_set:
        return 0
    return len(a_set & b_set) / len(a_set | b_set)

df_lemma = pd.read_csv("lemmatized.csv")
df_lemma["jaccard"] = df_lemma["lemmatized"].apply(lambda x: jaccard_similarity(giris_cumlesi, x))
top5_lemma = df_lemma.sort_values(by="jaccard", ascending=False).head(5)

df_stem = pd.read_csv("stemmed.csv")
df_stem["jaccard"] = df_stem["stemmed"].apply(lambda x: jaccard_similarity(giris_cumlesi, x))
top5_stem = df_stem.sort_values(by="jaccard", ascending=False).head(5)

jaccard_df = pd.DataFrame({
    "lemmatized_cumle": top5_lemma["lemmatized"].values,
    "lemmatized_jaccard": top5_lemma["jaccard"].values,
    "stemmed_cumle": top5_stem["stemmed"].values,
    "stemmed_jaccard": top5_stem["jaccard"].values,
})

jaccard_df.to_csv("jaccard_benzerlik_tablosu.csv", index=False)
```

```python
import pandas as pd

df = pd.read_excel("model_bazli_benzerlik_puanlari.xlsx")

cumleler = df["Unnamed: 1"].astype(str).str.split(" - ", expand=True)
cumleler = cumleler.reindex(columns=range(5))
cumleler.columns = ["Cümle 1", "Cümle 2", "Cümle 3", "Cümle 4", "Cümle 5"]

df = pd.concat([df[["Model"]], cumleler], axis=1)

model_cumleleri = {}
for _, row in df.iterrows():
    model = row["Model"]
    cumle_seti = {
        str(row["Cümle 1"]).strip(),
        str(row["Cümle 2"]).strip(),
        str(row["Cümle 3"]).strip(),
        str(row["Cümle 4"]).strip(),
        str(row["Cümle 5"]).strip()
    }
    model_cumleleri[model] = cumle_seti

modeller = list(model_cumleleri.keys())

jaccard_matrix = pd.DataFrame(index=modeller, columns=modeller)

for m1 in modeller:
    for m2 in modeller:
        set1 = model_cumleleri[m1]
        set2 = model_cumleleri[m2]
        kesisim = len(set1 & set2)
        birlesim = len(set1 | set2)
        jaccard = kesisim / birlesim if birlesim != 0 else 0
        jaccard_matrix.loc[m1, m2] = round(jaccard, 2)

jaccard_matrix = jaccard_matrix.astype(float)
jaccard_matrix.to_excel("model_model_jaccard_matrisi.xlsx")
```
